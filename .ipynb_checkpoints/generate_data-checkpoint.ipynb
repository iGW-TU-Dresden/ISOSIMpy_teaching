```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
df = pd.read_csv("model_input/atm_tracers.csv", sep=";")

# df['Date'] = df.Date.astype(np.datetime64)
# df['Date'] = df.Date.dt.year
# df.set_index('Date')

# df = df.drop([i for i in range(13)])
# df = df[~df.Date.str.contains(".")]
df = df[df.Date == df.Date // 1]
df.Date = df.Date.astype(int)

# df['Date'] = df.Date.astype(np.datetime64)
# df['Date'] = df.Date.dt.year
# df.set_index('Date')

index = pd.date_range(start="1940-01-15", periods=len(df), freq='A')
df.index = index

df.tail(20)
```

```python
df = df.drop(columns=["Date", "Unnamed: 5", "Unnamed: 6"])
df.head()
```

```python
df = df.resample("M").interpolate(method="linear")
# df.Index = df.index.apply(lambda x: x.strftime('%Y-%m'))

df.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))

p = np.array([0.95, 1., 0.7, 0.6, 0.2, 0.15, 0.1, 0.1, 0.2, 0.25, 0.7, 0.9])
p = np.tile(p, len(df)//12)
p = np.insert(p, 0, 1., axis=0)
p *= 100
np.random.seed(1)
p += np.random.normal(0, 20, size=(len(p)))

ax.plot(df.index, p)
ax.set_xticks(df.index[::12])
_ = ax.set_xticklabels(labels=df.index[::12], rotation=90)

# ax.set_xlim(datetime.date(1940, 12, 31), datetime.date(1950, 12, 31))
```

```python
print(p.mean())
```

```python
df["prec"] = p
df["Month"] = df.index.month
df["Year"] = df.index.year
df.head()
```

```python
df = df.set_index("Month")
```

```python
df = df[["Year", "prec", "CFC12", "CFC11", "CFC113", "SF6"]]
```
